***London Metal Exchange (LME)***

The London Metal Exchange is the world's largest market for industrial metals. Established in 1877, it provides a global marketplace for trading futures and options contracts for a wide range of base and other metals, including aluminum, copper, nickel, and zinc. The LME is the central hub for price discovery and risk management in the global metals industry. Its prices are used as the global benchmark. A unique feature of the LME is its system of physical delivery, where contracts can be settled with physical metal stored in LME-approved warehouses around the world.

***Japan Exchange Group (JPX)***

The Japan Exchange Group is the leading stock exchange group in Japan, formed through the 2013 merger of the Tokyo Stock Exchange (TSE) and the Osaka Securities Exchange (OSE). The JPX provides a comprehensive marketplace for a wide array of financial instruments. Its main components include the Tokyo Stock Exchange, which is home to the majority of Japan's largest public companies and offers trading in stocks, bonds, and exchange-traded funds (ETFs), and the Osaka Exchange, which specializes in derivatives products such as futures and options on major Japanese stock market indices like the Nikkei 225.

***US Stock Market***

The US stock market is the largest in the world by market capitalization and consists of several exchanges where shares of publicly traded companies are bought and sold. The two most prominent exchanges are the New York Stock Exchange (NYSE), known for its listings of large, well-established companies, and the Nasdaq Stock Market, which is recognized for its high concentration of technology and growth-oriented companies. The US stock market serves as a primary source of capital for businesses and a key avenue for individual and institutional investment. Its performance is often seen as an indicator of the health of the US and global economies.

***Foreign Exchange (Forex) Market***

The Foreign Exchange market, commonly known as Forex or FX, is a global, decentralized market for the trading of currencies. It is the largest and most liquid financial market in the world, with trillions of dollars traded daily. Unlike stock markets, the Forex market has no central physical location and operates 24 hours a day, five days a week through a network of banks, corporations, and individual traders. The primary purpose of the Forex market is to facilitate international trade and investment by enabling currency conversion. It also serves as a major platform for speculation on the fluctuating values of different currencies.

1. **US_Stock_EWJ_adj_close**


This refers to the adjusted closing price of the iShares MSCI Japan ETF (ticker symbol: EWJ), which is traded on a U.S. stock exchange.

US_Stock: The security is traded within the United States stock market.

EWJ: This is the ticker symbol for the iShares MSCI Japan ETF, a fund that tracks the performance of the Japanese stock market.

adj_close: This stands for "adjusted close," which is the closing price of the ETF after accounting for distributions like dividends and capital gains.


2. **FX_GBPCAD**


This represents the foreign exchange rate between the British Pound Sterling (GBP) and the Canadian Dollar (CAD).

FX: An abbreviation for Forex or Foreign Exchange.

GBPCAD: This is the currency pair. It tells you how many Canadian Dollars are needed to buy one British Pound.


3. **JPX_Gold_Rolling-Spot_Futures_Open**

This is the opening price for a Gold Rolling Spot Futures contract traded on the Japan Exchange Group (JPX).

JPX: The Japan Exchange Group, the main stock exchange in Japan.

Gold: The underlying commodity being traded.

Rolling-Spot_Futures: A type of futures contract that doesn't have a fixed expiry date and automatically rolls forward, mimicking a spot market price.

Open: Refers to the price at the beginning of the trading session.

4. **LME_AH_Close**

This indicates the official closing price for a Primary Aluminium High Grade contract on the London Metal Exchange (LME).

LME: The London Metal Exchange, the global center for industrial metals trading.

AH: The LME's ticker symbol for Primary Aluminium High Grade.

Close: The final price at the end of the trading day.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. SETUP YOUR DATA (as you described)
# One DataFrame, one constant 'x', multiple 'y' columns

# splendid how dictionary is now ur new dataframe

# The list of y-columns you want to plot
y_columns_to_plot = ['US_Stock_EWJ_adj_close', 'FX_GBPCAD', 'JPX_Gold_Rolling-Spot_Futures_Open', 'LME_AH_Close']
y_titles = [
    'unitedStates_stock_iSharesMSCIJapanETF_adjustedClose',
    'foreignExchange_britishPound_canadianDollar',
    'japanExchangeGroup_gold_rollingSpotFutures_open',
    'londonMetalExchange_aluminiumHighGrade_close'
]

# 2. CREATE THE 2x2 GRID
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# 3. LOOP THROUGH COLUMNS AND AXES
# We loop through the flattened axes
# The counter 'i' will help us pick the correct column name
for i, ax in enumerate(axes.flatten()):

    # Pick the column name for this turn of the loop
    current_y_column = y_columns_to_plot[i]
    titles = y_titles[i]

    # Plot the constant x against the current y column
    sns.lineplot(data=dataframe, x='date_id', y=current_y_column, ax=ax)
    ax.set_title(f'Plot of {titles}')

# 4. TIDY UP AND SHOW
plt.tight_layout()
plt.show()

# lowkey looking at us stock price , currencies exchange (forex) , metals price (lme ah) and japan exchange group (stock market of japan in general)
```
