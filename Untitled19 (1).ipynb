```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming 'merged_df' is your DataFrame
# and 'target_0' is the name of your target column.

# Ensure we're working with numeric columns only
numeric_df = merged_df.select_dtypes(include=np.number)

# 1. Calculate the full correlation matrix
correlation_matrix = numeric_df.corr()

# 2. Find the correlations with the target variable
target_corr = correlation_matrix['target_0']

# 3. Select the top 5 positive and bottom 5 negative correlations
# We drop 'target_0' itself from the list to avoid self-correlation
top_positive = target_corr.drop('target_0').nlargest(5)
top_negative = target_corr.nsmallest(5)

# 4. Combine these into a single list of the most important features
top_features = pd.concat([top_positive, top_negative])
top_feature_names = top_features.index.tolist() + ['target_0']

print("Top features selected for the heatmap:")
print(top_features)

# 5. Create a smaller correlation matrix with only these features
small_corr_matrix = numeric_df[top_feature_names].corr()

# 6. Plot the new, focused heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    small_corr_matrix,
    annot=True,        # This is the key: it writes the numbers on the map
    fmt='.2f',         # Format numbers to 2 decimal places
    cmap='coolwarm',   # Use a diverging colormap for better visibility
    linewidths=.5
)
plt.title('Correlation Heatmap of Top Features with Target')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor

# assuming target_0 is our target
features  = merged_df.drop(columns = ['date_id' , 'target_0'])
target = merged_df['target_0']

# fill the missing values with median
features = features.fillna(features.median())
target = target.fillna(target.median())

# training a simple model
model = LGBMRegressor(random_state = 42)
model.fit(features,target)

# get and display feature importance..
feature_importance = pd.DataFrame({
    'features': features.columns,
    'importance': model.feature_importances_
    }).sort_values('importance' , ascending = False)

# printing top 10 most imp feature according to the model
print(feature_importance.head(10))
```

```python
# the most important feature for target_0 was target 80

# Create a 1-day lag for your most important feature
merged_df['target_80_lag1'] = merged_df['target_80'].shift(1)

# Create a 3-day lag for another important feature
merged_df['target_37_lag3'] = merged_df['target_37'].shift(3)

# Create a 7-day rolling average of your top feature
merged_df['target_80_rolling_mean_7'] = merged_df['target_80'].shift(1).rolling(window=7).mean()

# Create a 14-day rolling standard deviation (volatility)
merged_df['target_80_rolling_std_14'] = merged_df['target_80'].shift(1).rolling(window=14).std()
```
